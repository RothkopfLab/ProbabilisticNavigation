```python
def get_startposition(tr):
    start = np.array([tr.iloc[0].x, tr.iloc[0].z])
    if not np.linalg.norm(startpositions[0]-start) > np.linalg.norm(startpositions[1]-start):
        return startpositions[0]
    else: 
        return startpositions[1]
    
    
def analyze_trajectory(tr):
    #calculate segment ends 
    segment_ends = tr.reset_index().groupby("segment")["index"].last().values
    endpoints = tr.loc[segment_ends.tolist()][['segment','x', 'z']]

    #true reponse
    true_response = [endpoints[endpoints.segment == 0].x.values[0], endpoints[endpoints.segment == 0].z.values[0]]
    
    #true target (what about jitter?)
    true_target = [endpoints[endpoints.segment == 1].x.values[0], endpoints[endpoints.segment == 1].z.values[0]]
    
    #true response start 
    true_response_start = [endpoints[endpoints.segment  == 3].x.values[0], endpoints[endpoints.segment == 3].z.values[0]]
    
    
    #true trial start 
    true_start = [tr.iloc[0].x, tr.iloc[0].z]
    
    return true_start, true_target, true_response_start, true_response 
    
def plot_idealized_trajectory(trajectory_id, start, jitterx = 0.0, jittery = 0.0, ax = None, get_landmarks = False, condition = 1):
    posts = [int(trajectory_id[0]),int(trajectory_id[1]),int(trajectory_id[2])]
    segment_ends = [0, post_id[posts[0]].copy(), post_id[posts[1]].copy(), post_id[posts[2]].copy()]
    
    if ax is None:
        fig = plt.figure(figsize = (7,10))
        ax = fig.add_subplot(111)
        ax.set_aspect('equal')
    else:
        fig = ax.figure
        
    ax.scatter(start[0],start[1], marker = 'D', s = 100, label = 'Start')
    
    
    segment_ends[1] += np.array([jitterx,jittery])
    
    
    for i,entry in enumerate(segment_ends[1:]):
        ax.scatter(entry[0],entry[1], label = 'Post: ' + str(i+1))

        
    #first segment
    dxdy = segment_ends[1] - start
    ax.arrow(start[0], start[1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'orange')

    #second segment
    dxdy = segment_ends[2] - segment_ends[1]
    ax.arrow(segment_ends[1][0], segment_ends[1][1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'orange')

    #third segment
    dxdy = segment_ends[3] - segment_ends[2]
    ax.arrow(segment_ends[2][0], segment_ends[2][1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'orange')
    
    #return paths 
    dxdy = segment_ends[1] - segment_ends[3]
    retrn = ax.arrow(segment_ends[3][0], segment_ends[3][1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'lightblue', ls = '--', label = 'Estimated homing')

    ax.set_xlim(-3.5,3.5)
    ax.set_ylim(-5,5)

    ax.set_xlabel('x(m)')
    ax.set_ylabel('y(m)')

    ax.grid()

    lm1 = ax.scatter(landmark1[0], landmark1[1], label = 'landmark1', marker = '*', s = 300)
    #lm2 = ax.scatter(landmark2[0], landmark2[1], label = 'landmark2', marker = '*', s = 300)
    #lm3 = ax.scatter(landmark3[0], landmark3[1], label = 'landmark3', marker = '*', s = 300)
    
    lms = [lm1]
    ##lms = [lm1,lm2,lm3]
    
    if condition == 2:
        for lm in lms:
            lm.set_visible(False)
            
    if condition == 4: 
        lm1.set_offsets(landmark1_rotated)
        #lm2.set_offsets(landmark2_rotated)
        #lm3.set_offsets(landmark3_rotated)
        
    
    if get_landmarks:
        return fig, ax, lm1, lm2, lm3 
    
    return fig, ax 
  

def plot_raw_trajectory(trajectory, response = None, ax = None, a = 1.0, c = 'black'):
    tr = trajectory[['x', 'z', 'segment']]
    
    if ax is None:
        fig = plt.figure(figsize = (6,10))
        ax = fig.add_subplot(111)
        ax.set_aspect('equal')
    else:
        fig = ax.figure
        
    if response == 'returnPath':
        tr = tr[tr.segment == 0]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c,alpha = a)
    
    if response == 'endpoint':
        tr = tr.iloc[-1]
        ax.scatter(tr.values[0], tr.values[1], label = 'raw trajectory', color = c, s = 15, alpha = a)
        
    if response == 'startpoint':
        tr = tr[tr.segment == 0].iloc[0]
        ax.scatter(tr.values[0], tr.values[1], label = 'raw trajectory', color = c, s = 5, alpha = a)
    
    if response == 'fullPath':
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    
    if response == '1stSegment':
        tr = tr[tr.segment == 1]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    if response == '1stSegmentEndPoint':
        tr = tr[tr.segment == 1]
        ax.scatter(tr.values[-1,0], tr.values[-1,1], label = 'raw trajectory', color = c, s = 5, alpha = a)
    
    if response == '2ndSegment':
        tr = tr[tr.segment == 2]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    if response == '2ndSegmentEndPoint':
        tr = tr[tr.segment == 2]
        ax.scatter(tr.values[-1,0], tr.values[-1,1], label = 'raw trajectory', color = c, s = 5, alpha = a)
        
    if response == '3rdSegment':
        tr = tr[tr.segment == 3]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    if response == '3rdSegmentEndPoint':
        tr = tr[tr.segment == 3]
        ax.scatter(tr.values[-1,0], tr.values[-1,1], label = 'raw trajectory', color = c, s = 5, alpha = a)
    
   
        
    return fig, ax 
        
        
def plot_trial(unique_t_id):
    tr = full_trajectory_data[full_trajectory_data.unique_trajectory_id == unique_t_id]
    trajectory_id = tr.trajectory_id.iloc[0]

    fig, ax = plot_idealized_trajectory(trajectory_id, get_startposition(tr))
    plot_raw_trajectory(tr,'fullPath', ax = ax)
    
    ax.legend(loc = 4)
    ax.set_xlabel('x (m)')
    ax.set_ylabel('y (m)')
    
    fig.suptitle('Unique T_ID: ' +  str(tr.unique_trajectory_id.iloc[0]) +'\n Day: ' + str(tr.day.iloc[0]) + '   Trial Number: ' + str(tr.trial.iloc[0]) + '   Trajectory_ID:' + trajectory_id + '\n Condition: ' + trial_types[tr.condition.iloc[0]]);
```

```python
def get_startposition(tr):
    start = np.array([tr.iloc[0].x, tr.iloc[0].z])
    if not np.linalg.norm(startpositions[0]-start) > np.linalg.norm(startpositions[1]-start):
        return startpositions[0]
    else: 
        return startpositions[1]
    
    
def analyze_trajectory(tr):
    #calculate segment ends 
    segment_ends = tr.reset_index().groupby("segment")["index"].last().values
    endpoints = tr.loc[segment_ends.tolist()][['segment','x', 'z']]

    #true reponse
    true_response = [endpoints[endpoints.segment == 0].x.values[0], endpoints[endpoints.segment == 0].z.values[0]]
    
    #true target (what about jitter?)
    true_target = [endpoints[endpoints.segment == 1].x.values[0], endpoints[endpoints.segment == 1].z.values[0]]
    
    #true response start 
    true_response_start = [endpoints[endpoints.segment  == 3].x.values[0], endpoints[endpoints.segment == 3].z.values[0]]
    
    
    #true trial start 
    true_start = [tr.iloc[0].x, tr.iloc[0].z]
    
    return true_start, true_target, true_response_start, true_response 
    
def plot_idealized_trajectory(trajectory_id, start, jitterx = 0.0, jittery = 0.0, ax = None, get_landmarks = False, condition = 1):
    posts = [int(trajectory_id[0]),int(trajectory_id[1]),int(trajectory_id[2])]
    segment_ends = [0, post_id[posts[0]].copy(), post_id[posts[1]].copy(), post_id[posts[2]].copy()]
    
    if ax is None:
        fig = plt.figure(figsize = (7,10))
        ax = fig.add_subplot(111)
        ax.set_aspect('equal')
    else:
        fig = ax.figure
        
    ax.scatter(start[0],start[1], marker = 'D', s = 100, label = 'Start')
    
    
    segment_ends[1] += np.array([jitterx,jittery])
    
    
    for i,entry in enumerate(segment_ends[1:]):
        ax.scatter(entry[0],entry[1], label = 'Post: ' + str(i+1))

        
    #first segment
    dxdy = segment_ends[1] - start
    ax.arrow(start[0], start[1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'orange')

    #second segment
    dxdy = segment_ends[2] - segment_ends[1]
    ax.arrow(segment_ends[1][0], segment_ends[1][1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'orange')

    #third segment
    dxdy = segment_ends[3] - segment_ends[2]
    ax.arrow(segment_ends[2][0], segment_ends[2][1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'orange')
    
    #return paths 
    dxdy = segment_ends[1] - segment_ends[3]
    retrn = ax.arrow(segment_ends[3][0], segment_ends[3][1], dxdy[0], dxdy[1], width = 0.025, length_includes_head = True, color = 'lightblue', ls = '--', label = 'Estimated homing')

    ax.set_xlim(-3.5,3.5)
    ax.set_ylim(-5,5)

    ax.set_xlabel('x(m)')
    ax.set_ylabel('y(m)')

    ax.grid()

    lm1 = ax.scatter(landmark1[0], landmark1[1], label = 'landmark1', marker = '*', s = 300)
    #lm2 = ax.scatter(landmark2[0], landmark2[1], label = 'landmark2', marker = '*', s = 300)
    #lm3 = ax.scatter(landmark3[0], landmark3[1], label = 'landmark3', marker = '*', s = 300)
    
    lms = [lm1]
    ##lms = [lm1,lm2,lm3]
    
    if condition == 2:
        for lm in lms:
            lm.set_visible(False)
            
    if condition == 4: 
        lm1.set_offsets(landmark1_rotated)
        #lm2.set_offsets(landmark2_rotated)
        #lm3.set_offsets(landmark3_rotated)
        
    
    if get_landmarks:
        return fig, ax, lm1, lm2, lm3 
    
    return fig, ax 
  

def plot_raw_trajectory(trajectory, response = None, ax = None, a = 1.0, c = 'black'):
    tr = trajectory[['x', 'z', 'segment']]
    
    if ax is None:
        fig = plt.figure(figsize = (6,10))
        ax = fig.add_subplot(111)
        ax.set_aspect('equal')
    else:
        fig = ax.figure
        
    if response == 'returnPath':
        tr = tr[tr.segment == 0]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c,alpha = a)
    
    if response == 'endpoint':
        tr = tr.iloc[-1]
        ax.scatter(tr.values[0], tr.values[1], label = 'raw trajectory', color = c, s = 15, alpha = a)
        
    if response == 'startpoint':
        tr = tr[tr.segment == 0].iloc[0]
        ax.scatter(tr.values[0], tr.values[1], label = 'raw trajectory', color = c, s = 5, alpha = a)
    
    if response == 'fullPath':
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    
    if response == '1stSegment':
        tr = tr[tr.segment == 1]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    if response == '1stSegmentEndPoint':
        tr = tr[tr.segment == 1]
        ax.scatter(tr.values[-1,0], tr.values[-1,1], label = 'raw trajectory', color = c, s = 5, alpha = a)
    
    if response == '2ndSegment':
        tr = tr[tr.segment == 2]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    if response == '2ndSegmentEndPoint':
        tr = tr[tr.segment == 2]
        ax.scatter(tr.values[-1,0], tr.values[-1,1], label = 'raw trajectory', color = c, s = 5, alpha = a)
        
    if response == '3rdSegment':
        tr = tr[tr.segment == 3]
        ax.plot(tr.values[:,0], tr.values[:,1], label = 'raw trajectory', color = c, alpha = a)
        
    if response == '3rdSegmentEndPoint':
        tr = tr[tr.segment == 3]
        ax.scatter(tr.values[-1,0], tr.values[-1,1], label = 'raw trajectory', color = c, s = 5, alpha = a)
    
   
        
    return fig, ax 
        
        
def plot_trial(unique_t_id):
    tr = full_trajectory_data[full_trajectory_data.unique_trajectory_id == unique_t_id]
    trajectory_id = tr.trajectory_id.iloc[0]

    fig, ax = plot_idealized_trajectory(trajectory_id, get_startposition(tr))
    plot_raw_trajectory(tr,'fullPath', ax = ax)
    
    ax.legend(loc = 4)
    ax.set_xlabel('x (m)')
    ax.set_ylabel('y (m)')
    
    fig.suptitle('Unique T_ID: ' +  str(tr.unique_trajectory_id.iloc[0]) +'\n Day: ' + str(tr.day.iloc[0]) + '   Trial Number: ' + str(tr.trial.iloc[0]) + '   Trajectory_ID:' + trajectory_id + '\n Condition: ' + trial_types[tr.condition.iloc[0]]);
```

```python
def rotate(origin, point, angle):
    """
    Rotate a point counterclockwise by a given angle around a given origin.

    The angle should be given in radians.
    """
    ox, oy = origin
    px, py = point

    qx = ox + math.cos(angle) * (px - ox) - math.sin(angle) * (py - oy)
    qy = oy + math.sin(angle) * (px - ox) + math.cos(angle) * (py - oy)
    
    return qx, qy
```

```python
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
from IPython.display import HTML

def get_trajectory_animation(coords, goal = None, dt = 1/50, showFov = None,fov = 60, cone_len = 20, save = False):
    #calculate time 
    timestamps = np.arange(0,len(coords), dt)
    
    #setup animation
    t = len(coords)
    animation_buffer = []

    fig, ax = plt.subplots(figsize = [7,10])
    ax.set_xlim(-3.5,3.5)
    ax.set_ylim(-5,5)
        
    position, e = plot_positional_belief(coords[0],0.5 * np.eye(2)) 
    
    line, = ax.plot(coords[0,0], coords[0,0], color='black', ls = 'dotted')

    text_i = ax.text(s = 'iteration:' + str(0), x = 0.65, y = 0.85, transform = fig.transFigure)
    text_t = ax.text(s = 'animation time:' + str(0), x = 0.65, y = 0.825, transform = fig.transFigure)
    
    if showFov:
        left, right = plot_agent_fov(coords[0].reshape(3,1), fov, cone_len)
        
    if goal is not None:
        ax.plot((coords[0][0],goal[0]), (coords[0][1],goal[1]))
        ax.scatter(np.array(goal[0]), np.array(goal[1]))
        
    def animate(i):
        position.set_offsets(coords[i][:2])
        e.set_center(coords[i][:2])
        text_i.set_text('iteration: ' + str(i))
        text_t.set_text('animation time: {:.2f} s'.format(i*107 / 1000))

        line.set_data(coords[:i,0],coords[:i,1])
        
        if showFov:
            update_fov(left,right,coords[i].reshape(3,1), fov, cone_len)
            
            
    anim = FuncAnimation(
        fig, animate, interval=127, frames=t)
    
    if save:
        return anim

    return HTML(anim.to_html5_video())
```

```python
def get_heading(point1,point2):
    d = point2 - point1
    heading_angle = np.arctan2(d[1], d[0]).reshape(1)
    
    return heading_angle

def get_heading_angles_over_time(coords):
    heading_angles = []
    
    for i in range(0,len(coords)-1):
        heading = np.rad2deg(get_heading(coords[i,:], coords[i+1,:]))
        heading_angles.append(heading[0])
    
    return heading_angles
```

```python
def optimize_hyperparams_BO(x_cost,y_cost, theta_cost, u_cost, w_cost, uw_cost):
    route = get_route('148')
    mean_heading_angle = np.rad2deg(get_heading(route[1], route[2])[0])

    #weight matrix for state cost
    Q = np.zeros((3,3))
    Q[0,0] = x_cost
    Q[1,1] = y_cost
    Q[2,2] = theta_cost

    #weight matrix for control cost 
    R = np.zeros((2,2))
    R[0,0] = u_cost
    R[1,1] = w_cost
    R[0,1] = uw_cost
    R[1,0] = uw_cost

    start = vertcat(-1.80314,0.420974, -2.62859)
    end = vertcat(1.09866, -0.0365229, -0.154009)

    x, u = mpc_multiple_shooting(
        start = start, 
        end = end, 
        dt = 1/10, 
        prediction_horizon = 100, 
        simulation_time = 20, 
        v_max = 2.00,
        v_min = 0.0,
        w_max = pi/2,
        w_min = -pi/2, 
        R = R,
        Q = Q,
        alpha = False)


    heading = np.array(get_heading_angles_over_time(x))
    cost = np.var(heading) + np.mean((heading-mean_heading_angle)**2) + np.sum((x[-1][:2] - route[2])**2)
    
    return -cost
```

```python
from bayes_opt import BayesianOptimization

# Bounded region of parameter space
pbounds = {
    'x_cost' : (0,100) ,
    'y_cost' : (0,100), 
    'theta_cost' : (0,100), 
    'u_cost' : (0,100), 
    'w_cost': (0,100), 
    'uw_cost' : (0,100)}

optimizer = BayesianOptimization(
    f=optimize_hyperparams_BO,
    pbounds=pbounds,
    random_state=1
)

optimizer.maximize(init_points = 100, n_iter = 1000)
```

```python
def get_R_and_Q(params):
    #weight matrix for state cost
    Q = np.zeros((3,3))
    Q[0,0] = params['x_cost']
    Q[1,1] = params['y_cost']
    Q[2,2] = params['theta_cost']

    #weight matrix for control cost 
    R = np.zeros((2,2))
    R[0,0] = params['u_cost']
    R[1,1] = params['w_cost']
    R[0,1] = params['uw_cost']
    R[1,0] = params['uw_cost']
    
    return R,Q 
    
R,Q = get_R_and_Q(optimizer.res[44]['params'])

start = vertcat(-1.80314,0.420974, -2.62859)
end = vertcat(1.09866, -0.0365229, -0.154009)

x, u = mpc_multiple_shooting(
    start = start, 
    end = end, 
    dt = 1/50,
    prediction_horizon = 200, 
    simulation_time = 20, 
    v_max = 2.00,
    v_min = 0.0,
    w_max = pi/2,
    w_min = -pi/2, 
    R = R,
    Q = Q,
    alpha = False)

get_trajectory_animation(x, dt = 1/50, goal = (end[0],end[1]), showFov=True)
```

```python
def get_dataframe(full_x, dt = 1/50):
    full_x = np.concatenate([x for x in full_x])
    #convert into a trajectory 
    df = pd.DataFrame(columns=['VPCode','trajectory_id','unique_trajectory_id','trial','condition','segment','timing','x','z','angle','day'])
    df.x = full_x[:,0]
    df.z = full_x[:,1]
    df.angle = full_x[:,2]
    df.trial = 1
    df.condition = 1 
    df.segment = 1 #TODO this goes better :D 
    df.timing = np.linspace(0,full_x.shape[0] * dt,full_x.shape[0])
    df.trajectory_id = '148'
    df.VPCode = 'Simulation'
    df.day = 1 
    df.unique_trajectory_id = 1
    
    return df 

def optimize_hyperparams_BO(x_cost,y_cost, theta_cost, alpha):
    route = get_route('148')
    
    #weight matrix for state cost
    Q = np.zeros((3,3))
    Q[0,0] = x_cost
    Q[1,1] = y_cost
    Q[2,2] = theta_cost

    #weight matrix for control cost 
    R = np.zeros((2,2))
    R[0,0] = alpha * 1
    R[1,1] = 1
    R[0,1] = 0
    R[1,0] = 0
    
    #start at zero cost 
    total_cost = 0
    
    #initialize first path
    starting_angle = np.array(np.deg2rad(-145)).reshape(1,)
    start = DM(np.concatenate((route[0], starting_angle)))

    #get heading angle 
    d = route[1] - route[0]
    heading_angle = np.arctan2(d[1], d[0]).reshape(1)
    
    #end point 
    end = DM(np.concatenate((route[1], heading_angle)))
    
    full_x = []

    ##############################
            #Simulation# 
    ##############################
    for k in np.arange(2,6): 
        x, u = mpc_multiple_shooting(
                                 start = start, 
                                 end = end, 
                                 dt = 1/50, 
                                 prediction_horizon = 250, 
                                 simulation_time = 7.5, 
                                 v_max = 2.00,
                                 v_min = -0.5,
                                 w_max = pi,
                                 w_min = -pi, 
                                 R = R,
                                 Q = Q,
                                 alpha = None)
        
        
        full_x.append(x)

        #cost
        mean_heading_angle = np.rad2deg(heading_angle[0])
        heading = np.array(get_heading_angles_over_time(x))
        cost = np.var(heading) + np.mean((heading-mean_heading_angle)**2) + np.sum((x[-1] - np.array(end)**2))
        
        #add to total_Cost                                                                        
        total_cost = total_cost + cost 

        x[-1,2] = normalize_angle(x[-1,2])

        #starting point after trajectory is the final position
        start = DM(x[-1,:])
                                                                                   
        #figure out heading direction
        if k < 5: 
            end   = route[k]
            d = np.array(end) - np.array(start)[:2].reshape(2)
            heading_angle = np.arctan2(d[1], d[0]).reshape(1)
            end = DM(np.concatenate((route[k], heading_angle)))
            
            
    df = get_dataframe(full_x)
    
    anim = get_trajectory_animation(df, showFov=True, save = True)
    
    fname = str(datetime.datetime.now()).replace(':','_').replace('.','_')[:-7] + "_" + "{:.0f}".format(-total_cost)
    #write to file 
    Writer = animation.writers['ffmpeg']
    writer = Writer(fps=50, metadata=dict(R = str(R.flatten()), Q = str(Q.flatten())), bitrate=1800)
    anim.save('animations/cost_optimization/' + fname  + '.mp4', writer=writer)
    
    plt.close()
                
    return -total_cost
```

```python
from bayes_opt import BayesianOptimization

# Bounded region of parameter space
pbounds = {
    'x_cost' : (0,100) ,
    'y_cost' : (0,100), 
    'theta_cost' : (0,100), 
    'alpha' : (0,100)}

optimizer = BayesianOptimization(
    f=optimize_hyperparams_BO,
    pbounds=pbounds,
    random_state=1337
)

optimizer.maximize(init_points=50, n_iter = 500)
```
