# Do MPC Version 

```python
# import sys
sys.path.append("..")

from parameters import get_params
from mpc import * 


#initialize experimental parameters
experiment = 'nardini2008'


if experiment == 'nardini2008':
    params = get_params(experiment_name = 'nardini2008', trajectory_id = 'random', n_landmarks = 3, target_jitter=False)
    
if experiment == 'chen2017':
    params = get_params(experiment_name = 'chen2017', trajectory_id = 'random', n_landmarks = 3, start_position = [0,-2], target_jitter=True)
    params['start_position'] = [0,-2]
    
if experiment == 'zhao2015':
    params = get_params(experiment_name = 'zhao2015a', trajectory_id = 'random', n_landmarks = 3, target_jitter=False)


    

#choose experimental condition  
params['condition'] = 'combined' #self-motion,conflict,landmark
params['fov'] = 100

#create a model of the agen<t 
model = Model(n_landmarks = len(params['landmark_locations'] + params['pickup_locations'])//2,
              motion_limits = params['motion_limits'],
              R_diag = params['observation_cov'], 
              Q_diag = params['motion_cov'],
              R_diag_generative = params['observation_cov'],
              Q_diag_generative = params['motion_cov'], 
              V = 0.01295**2,
              V_generative = 0.01295**2,
              representation_type = 'all',
              fov = params['fov'])

#create planner 
planner = BeliefSpacePlanner(model, 
                             N = params['planning_horizon'],
                             verbose = False)

solver, opts = planner.create_solver(solver = 'ma57', 
                                     BFGS = True, 
                                     warm_start = True)

live_plot = Plotting()

x_true, mu_t, Sigma_t, obs0, obs_state, obs_belief, p, args, task, logger  = init_state(params, model, planner, solver)


mpc_iter = 0
sol = None

params['arena_limits'] = ((-8, 8), (-8, 8))


mpc = MPC(model, solver, planner, sol, params, obs0, obs_state, obs_belief, p, args, task, logger, live_plot)
```

```python
%matplotlib notebook
mpc_iter = 0
mpc.planner.reinit=True
n_iter = 0

while True:
    x_true, mu_t, Sigma_t, mpc_iter = mpc.update(x_true, mu_t, Sigma_t, mpc_iter, verbose = True)
    
    if task.task_finished:
        n_iter += 1         
        if n_iter > 10:
            break  
```
